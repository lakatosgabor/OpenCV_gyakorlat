# Hisztogram széthúzás feladat (HF)
A 18. sorban definiáljuk az np.uint8 függvény paraméterének a hisztogram széthúzás képletét.
A kódban, a képlesz számításához szükséges változók rendelkezésre állnak:


*   input_image
*   min_intensity
*   desired_max_intensity
*   desired_min_intensity
*   max_intensity
*   min_intensity

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

# Kép betöltése
input_image = cv.imread("/content/drive/MyDrive/opencv/images/lena.jpg", cv.IMREAD_GRAYSCALE)

# Minimum és maximum pixelértékek kiszámítása
min_intensity = np.min(input_image)
max_intensity = np.max(input_image)

# Kívánt minimum és maximum értékek
desired_min_intensity = 0
desired_max_intensity = 255

# Hisztogram széthúzás számítás
stretched_image = np.uint8()

# Eredeti kép hisztogramja
hist_original = cv.calcHist([input_image], [0], None, [256], [0, 256])

# Széthúzott kép hisztogramja
hist_stretched = cv.calcHist([stretched_image], [0], None, [256], [0, 256])

f, axs = plt.subplots(2, 2, figsize=(30, 10))

plt.subplot(221), plt.imshow(cv.cvtColor(input_image, cv.COLOR_BGR2RGB))
plt.subplot(222), plt.imshow(cv.cvtColor(stretched_image, cv.COLOR_BGR2RGB))
plt.subplot(223), plt.plot(hist_original)
plt.subplot(224), plt.plot(hist_stretched)
```
