# Kép betöltése és megjelenítése

````python
# Fájl elérési útja a Colab fájlrendszerében, módosítsd a fájl elérési útját a saját mappád és fájlod nevére
imageTennisPath = "/content/photos/tennis.jpg"
imageBusPath = "/content/photos/bus.jpg"

# Kép beolvasása
imageTennis = cv2.imread(imageTennisPath)
imageBus = cv2.imread(imageBusPath)

if imageTennisPath is not None:
    # Kép megjelenítése
    cv2_imshow(imageTennis)
else:
    print("Hiba: Nem sikerült beolvasni a képet.")
````

# Hisztogram
A hisztogram egy olyan grafikus ábrázolás, amely egy kép szürkeségi értékeinek eloszlását mutatja be. Az X tengelyen általában a szürkeségi értékek tartománya található (általában 0-tól 255-ig terjed, ahol 0 a teljesen fekete, 255 pedig a teljesen fehér), míg az Y tengelyen a képen található pixelek mennyiségét vagy gyakoriságát mutatja.

````python
# Kép hisztrogramjának számítása
histogram = cv2.calcHist([imageTennis], [0], None, [256], [0, 256])

# Hisztrogram megjelenítése
plt.figure(figsize=(8, 5))
plt.title("Kép Hisztrogram")
plt.xlabel("Szürkeségi érték")
plt.ylabel("Mennyiség")
plt.plot(histogram)
plt.xlim([0, 256])
plt.show()
````
